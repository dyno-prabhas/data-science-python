```python
import pandas as pd
```

```python
df = pd.read_csv('stock_data.csv', header=None) ### it doesn't take columns names which are exists in csv files
df
```

```python
df = pd.read_csv('stock_data.csv', header=None, names=['ticker', 'eps', 'revenue', 'price', 'people'])
df
```

```python
df = pd.read_csv('stock_data.csv', nrows=3)
df
```

```python
df = pd.read_csv('stock_data.csv', na_values=["not available", "n.a."])
df
```

```python
df = pd.read_csv('stock_data.csv', na_values={
    'eps': ["not available", "n.a."],
    'revenue': ["not available", "n.a.", -1],
    'people': ["not available", "n.a."]
})
df
```

```python
def convert_people_cell(cell):

    if cell == "n.a.":
        return "Sam Walton"
    return cell

df = pd.read_csv('stock_data.csv', converters={
    'people': convert_people_cell
})
df
```
